```python
from xmldiff import main
import difflib
from xml.etree.ElementTree import canonicalize
import os
import pandas as pd
```

```python
files = [
 '2_op13_3',
 '1_op12_3',
 'FP_op81_3',
 '4_op44ii_1',
 '3_op44i_1',
 '5_op44iii_4',
 '1_op12_4',
 'FP_op81_1',
 'FP_op81_2',
 '2_op13_4',
 '1_op12_1',
 '3_op44i_3',
 '4_op44ii_4',
 '3_op44i_2',
 '4_op44ii_2',
 '5_op44iii_1',
 '5_op44iii_2',
 '5_op44iii_3',
 'FP_op81_4',
 '3_op44i_4',
 '4_op44ii_3',
 '2_op13_2',
 '1_op12_2',
 '2_op13_1'
]

versions = [
    'omr',
    'corrected',
    'revised',
    'aligned'
]
```

```python
folder = 'all_files'
df_dict = {}
for file in files:
    row_dict = {}
    C = 0
    for v1, v2 in zip(versions, versions[1:]):
        f1 = '{}_{}.musicxml'.format(file, v1)
        f2 = '{}_{}.musicxml'.format(file, v2)
        filepath1 = os.path.join(folder, f1)
        filepath2 = os.path.join(folder, f2)        
        with open(filepath1) as fd:            
            data1 = canonicalize(fd.read(), strip_text=True)
        with open(filepath2) as fd:                        
            data2 = canonicalize(fd.read(), strip_text=True)
        sm = difflib.SequenceMatcher(None, data1, data2)
        print(f'{f1} vs {f2}', sm.real_quick_ratio())
        row_dict[f'C{C}_to_C{C+1}'] = sm.real_quick_ratio()
        C += 1
    df_dict[file] = row_dict
```

```python
df = pd.DataFrame(df_dict).transpose()
```

```python
for col in df.columns:
    print(col)
    print(f'{df[col].median()} [{df[col].min()} {df[col].max()}]')
df
```
